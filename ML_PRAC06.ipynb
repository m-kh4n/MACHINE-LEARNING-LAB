# Aim : Implement a neural network for MLP Digit-Classifier using TensorFlow




# Executions :
##1. Importing Libraries and Datasets


````python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Here we are using the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalization and Label Encoding

train_images = train_images / 255.0
test_images = test_images / 255.0
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
````

##2. Creation of the MLP Model

1.  Initialize weights and biases for each layer, typically using small random values.
2.  Perform feedforward propagation by applying activation functions to the weighted sum of inputs for each neuron in hidden layers.
3.  Compute the loss between predicted and actual values using a suitable loss function (e.g., cross-entropy for classification, mean squared error for regression).
4.  Backpropagate the error by computing gradients of the loss with respect to weights and biases and update parameters using gradient descent or other optimization algorithms.
5.  Iteratively repeat steps 2-4 for a specified number of epochs, adjusting parameters to minimize the loss.
6.  Evaluate the model on a separate validation set to monitor generalization performance during training.
7.  Assess the final model performance on a test set not used during training to estimate its ability to generalize to new, unseen data.
8.  Optionally perform hyperparameter tuning and model fine-tuning to improve performance.

````python
model = models.Sequential()

# Flatten the input data (28x28 images) to a 1D array
model.add(layers.Flatten(input_shape=(28, 28)))

# Add a fully connected hidden layer with 128 units and ReLU activation
model.add(layers.Dense(128, activation='relu'))

# Add another fully connected hidden layer with 64 units and ReLU activation
model.add(layers.Dense(64, activation='relu'))

# Add the output layer with 10 units (one for each digit) and softmax activation
model.add(layers.Dense(10, activation='softmax'))

#compling the MLP modal
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
````

##3. Training Testing and Evaluating the Model

````python
history=model.fit(train_images, train_labels, epochs=10, batch_size=64, validation_data=(test_images, test_labels))

test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc * 100}%')

training_results = history

train_loss = training_results.history['loss']
train_acc = training_results.history['accuracy']
valid_loss = training_results.history['val_loss']
valid_acc = training_results.history['val_accuracy']
````

##4. Result Visualization

````python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(train_loss, label="Training Loss", color="g")
plt.plot(valid_loss, label="Validation Loss", color="b")
plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(train_acc, label="Training Accuracy", color="g")
plt.plot(valid_acc, label="Validation Accuracy", color="b")
plt.title("Training and Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()

plt.show()
````

#Result Analysis :

*   **Performance Insights :** Machine learning model results provide valuable insights into dataset performance.
*   **MLP Accuracy :** The trained Multi-Layer Perceptron achieves a notable 97% accuracy on the MNIST dataset, showcasing its ability to recognize handwritten digits effectively.
*   **Generalization Strength :** The model demonstrates strong generalization, reflected in the close alignment of validation accuracy with training accuracy. This indicates suitability for real-world applications without overfitting.

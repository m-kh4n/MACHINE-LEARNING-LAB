###Import TensorFlow Libraries

````python
# Importing peripheral libraries

import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
````

````python
# Importing TensorFlow libraries
# pip install tensorflow

import tensorflow as tf
print(tf.__version__)
````

````python
# Constants and arithmetics using tensorflow

tensor_a = tf.constant(5.0)
tensor_b = tf.constant(3.2)
tensor_c = tf.constant(2.5)

print(tf.add(tensor_a, tensor_b))
print(tf.subtract(tensor_b, tensor_c))
print(tf.multiply(tensor_c, tensor_a))
print(tf.divide(tensor_a, tensor_c))
````

````python
# Graphs

graph = tf.Graph()

with graph.as_default():
  tensor_a = tf.constant(3.0)
  tensor_b = tf.constant(4.2)
  result = tf.add(tensor_a, tensor_b)
  print(graph)
````

````python
# Datatypes in TensorFlow


# Scalars
scalar_int32 = tf.constant(5, dtype=tf.int32)
scalar_float32 = tf.constant(2.5, dtype=tf.float32)

# Vectors
vector_int32 = tf.constant([1, 2, 3], dtype=tf.int32)
vector_float64 = tf.constant([2.5, 3.0, 4.5], dtype=tf.float64)

# Matrices
matrix_int32 = tf.constant([[1, 2], [3, 4]], dtype=tf.int32)
matrix_float32 = tf.constant([[2.5, 3.0], [4.0, 5.5]], dtype=tf.float32)

# Print data types and values
print("Scalar (int32):", scalar_int32.numpy(), "Data Type:", scalar_int32.dtype)
print("Scalar (float32):", scalar_float32.numpy(), "Data Type:", scalar_float32.dtype)
print("\nVector (int32):", vector_int32.numpy(), "Data Type:", vector_int32.dtype)
print("Vector (float64):", vector_float64.numpy(), "Data Type:", vector_float64.dtype)
print("\nMatrix (int32):")
print(matrix_int32.numpy(), "Data Type:", matrix_int32.dtype)
print("Matrix (float32):")
print(matrix_float32.numpy(), "Data Type:", matrix_float32.dtype)
````

#Result :

As we conducted creation of constants and performed computations on them, we learned some things, that TensorFlow:

*   Allows dynamic computation with eager execution, providing immediate feedback and simplifying debugging.
*   Represents computations as static dataflow graphs, enabling optimization and parallel execution across devices.
*   Handles constants and tensors efficiently, supporting various data types and facilitating seamless integration with hardware accelerators.
